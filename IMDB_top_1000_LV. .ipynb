**Multi Index**

```python
outside = ['M1', 'M1', 'M1', 'M2', 'M2', 'M2','M3', 'M3', 'M3'] # zip ile list oluÅŸtuturma
inside = [1, 2, 3, 1, 2, 3, 5, 6, 7]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index = pd.MultiIndex.from_tuples(multi_index) # pandas series oluÅŸturma
hier_index
```

```python
np.random.seed(101)
df=pd.DataFrame(np.random.randn(9,4), index = hier_index, columns = ["A","B","C","D"]) # Pandas Multi Index OluÅŸturma
df
```

```python
df.index.names #df indexleri kapalÄ± durumda.. (dÄ±ÅŸ index (M1..) (iÃ§ index(1...)
```

```python
df.index.names = ["group", "num"] # df iÃ§ ve dÄ±ÅŸ index' isim verildi
df
```

```python
df.index # index'ler gÃ¶rÃ¼nÃ¼yor..
```

```python
df.index.levels # iÃ§ ve dÄ±ÅŸ indexler
```

```python
df.loc["M1"] # dÄ±ÅŸ(group) index ile seÃ§ileni getirme
```

```python
df.loc[("M1",2)] #dÄ±ÅŸ index M1 satÄ±r index = 2
```

```python
df.loc[("M1",2)][["A","D"]] # M1 = group(dÄ±ÅŸ) index , 2 = satÄ±r index, ["A", "D"] = sÃ¼tun adlarÄ±
```

```python
df.loc[[("M1",2)], "A" : "C"] # M1 = group(dÄ±ÅŸ) index , 2 = satÄ±r index, ["A": "D"] = sÃ¼tun adlarÄ±
```

```python
df.loc[("M1,3"):("M2")] # M1, M2 = dÄ±ÅŸ indek, 3 = satÄ±r index --> M1'in 3. satÄ±rÄ±ndan baÅŸla M2'nin tÃ¼m sayÄ±rlarÄ±nÄ± getir
```
