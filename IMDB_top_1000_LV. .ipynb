**Conditional Selection**

```python
df > 10 # condation için geçerli değerleri "True" geçersiz değerleri "False" olarak getir.(10'dan büyük değerler True)
```

```python
df[df > 10] # condation için geçerli değerleri getirir, geçersiz değerleri "NaN" olarak getir.(10'dan küçük değerler NaN)
```

```python
df[df["var2"] > 10] #seçilen sütun için condation'da geçerli değerleri getirdi. seçilen sütunda condation için geçersiz değerler DataFrame yok.diğer sütunlarda condation devre dışı, çalışmadı
```

```python
df[df["var1"] > 10][["var1"]] #condation için geçerli değerler ve seçilen sütun getirildi.
```

```python
df[(df["var1"] > 7) & (df["var1"] < 20)] # çoklu condition kullanımı
```

```python
df[(df["var1"] > 7) & (df["var1"] < 20)][["var1","var2"]] # çoklu condition kullanımı ve istenilen sütun/lar getirme
```

```python
df.loc[["a","f"], ["var1","var3"]] #loc ile satır ve sütun getirme
```

```python
df.loc[(df["var1"] > 15) | (df["var2"] >= 20)] #loc ile çoklu condition kullanımı
```

**Multi Index**

```python
outside = ['M1', 'M1', 'M1', 'M2', 'M2', 'M2','M3', 'M3', 'M3'] # zip ile list oluştuturma
inside = [1, 2, 3, 1, 2, 3, 5, 6, 7]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index = pd.MultiIndex.from_tuples(multi_index) # pandas series oluşturma
hier_index
```

```python
np.random.seed(101)
df=pd.DataFrame(np.random.randn(9,4), index = hier_index, columns = ["A","B","C","D"]) # Pandas Multi Index Oluşturma
df
```

```python
df.index.names #df indexleri kapalı durumda.. (dış index (M1..) (iç index(1...)
```

```python
df.index.names = ["group", "num"] # df iç ve dış index' isim verildi
df
```

```python
df.index # index'ler görünüyor..
```

```python
df.index.levels # iç ve dış indexler
```

```python
df.loc["M1"] # dış(group) index ile seçileni getirme
```

```python
df.loc[("M1",2)] #dış index M1 satır index = 2
```

```python
df.loc[("M1",2)][["A","D"]] # M1 = group(dış) index , 2 = satır index, ["A", "D"] = sütun adları
```

```python
df.loc[[("M1",2)], "A" : "C"] # M1 = group(dış) index , 2 = satır index, ["A": "D"] = sütun adları
```

```python
df.loc["M1":"M2"]
```

```python
df.loc[("M1",2):("M2",1)] # M1, M2 = dış indek, 2,1 = satır index --> M1'in 2. satırından başla M2'nin 1 satırına kadar(dahil) getir
```

```python
df.loc[("M1,3"):("M2")] # M1, M2 = dış indek, 3 = satır index --> M1'in 3. satırından başla M2'nin tüm sayırlarını getir
```

```python
df.loc[[("M1", 2),("M2",1)]] # M1'in 2. satır index ve M2'nin 1. satır indexlerinde olan değerleri getir
```

```python
df.loc[[("M1", 2),("M2",1)]]["A"] # A sütununda buluna M1'in 2. satır index ve M2'nin 1. satır indexlerinde olan değerleri getir
```
